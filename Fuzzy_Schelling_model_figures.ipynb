```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['default'])
```

betabinomial distribution:
uniform: a=1, b=1,
bimodal: a=0.5, b=0.5,
unimodal: a=2, b=2,
biased: a=2, b=1

```python
x = np.linspace(0,1,100)
for a,b in zip([1,0.5,2], [1,0.5,2]):
    beta_pdf = beta.pdf(x, a, b)
    plt.plot(x,beta_pdf, label= r'$ \alpha $' + ' = {}, '.format(a) +  r'$ \beta $' + ' = {}'.format(b))
plt.xlim([0,1])
plt.xlabel('x')
plt.ylabel('p.d.f. of x')
plt.legend(loc='upper center')
# plt.savefig('TeX/beta_dist.pdf',bbox_inches='tight')
plt.show();
```

```python
x = np.linspace(0,1,100)
for a,b,l in zip([1,0.5,2], [1,0.5,2], ['-', '--', '-.']): # ':'
    beta_pdf = beta.pdf(x, a, b)
    plt.plot(x,beta_pdf, label= r'$ \alpha $' + ' = {}, '.format(a) +  r'$ \beta $' + ' = {}'.format(b), linestyle=l)
plt.xlim([0,1])
plt.xlabel('x')
plt.ylabel('p.d.f. of x')
plt.legend(loc='best')
plt.savefig('TeX/beta_dist.pdf',bbox_inches='tight')
plt.show();
```

```python
ex1 = np.array([[np.nan, 0.2, 0.4],
                [0.7, 0.6, 0.9],
                [1, np.nan, 0.5]])
ex2 = np.array([[np.nan, 0.2, 0.4],
                [0.3, 0.6, 0.9],
                [0, np.nan, 0.7]])
NUM_COL = 2
NUM_ROW = 2
```

```python
# Change of mu distribution
fig, ax = plt.subplots(ncols=2)

left = ax[0].imshow(ex1, vmin=0, vmax=1, cmap="jet")
ax[0].set_title("stay (similarity % = 0.5)")# p = 0.5
ax[0].axes.get_xaxis().set_visible(False)
ax[0].axes.get_yaxis().set_visible(False)

for i in range(ex1.shape[0]):
    for j in range(ex1.shape[1]):
        if not np.isnan(ex1[i, j]):
            text = ax[0].text(j, i, ex1[i, j], ha="center", va="center", color="grey")

right = ax[1].imshow(ex2, vmin=0, vmax=1, cmap="jet")
ax[1].set_title("move (similarity % = 0.33)")
ax[1].axes.get_xaxis().set_visible(False)
ax[1].axes.get_yaxis().set_visible(False)

for i in range(ex2.shape[0]):
    for j in range(ex2.shape[1]):
        if not np.isnan(ex2[i, j]):
            text = ax[1].text(j, i, ex2[i, j], ha="center", va="center", color="grey")

axpos = ax[1].get_position()
cbar_ax = fig.add_axes([0.87, axpos.y0, 0.02, axpos.height])

cbar = fig.colorbar(right, cax=cbar_ax)
cbar.ax.set_title(r'$ \mu $')
plt.subplots_adjust(right=0.85)
plt.subplots_adjust(wspace=0.1)
# plt.savefig('TeX/Moore_neighbor_ex.pdf',bbox_inches='tight')
plt.show();
```
